```python
import pandas as pd
data = pd.read_excel('Orders датасет.xlsx', sheet_name='Orders Data')
data
```

```python
data.info()
```

```python
data['Order Date'] = data['Order Date'].fillna(data['Order Date'].mode()[0])
data['Quantity of Items'] = data['Quantity of Items'].fillna(data['Quantity of Items'].median())
data['Order Date'] = pd.to_datetime(data['Order Date'])

numerical_columns = [
    'Quantity of Items', 'Order Amount', 'Delivery Time Taken (mins)', 
    'Customer Rating-Food', 'Customer Rating-Delivery'
]
for column in numerical_columns:
    data[column] = pd.to_numeric(data[column], errors='coerce')

def handle_outliers(df, numerical_columns):
    for column in numerical_columns:
        Q1 = df[column].quantile(0.25)
        Q3 = df[column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        df[column] = df[column].clip(lower=lower_bound, upper=upper_bound)
    return df
data = handle_outliers(data, numerical_columns)
data = data.drop_duplicates()
data.to_excel('Cleaned_Orders_Data.xlsx', index=False)
data
```

```python
#Разведочный анализ данных (EDA) (30% от содержания):
numerical_columns = [
    'Quantity of Items', 'Order Amount', 'Delivery Time Taken (mins)', 
    'Customer Rating-Food', 'Customer Rating-Delivery'
]

numerical_stats = data[numerical_columns].describe().T
numerical_stats['median'] = data[numerical_columns].median()
numerical_stats['mode'] = data[numerical_columns].mode().iloc[0]

print("Descriptive Statistics:")
print(numerical_stats)
```

```python
#Разведочный анализ данных (EDA) (30% от содержания):
payment_mode_counts = data['Payment Mode'].value_counts()

payment_mode_sums = data.groupby('Payment Mode')['Order Amount'].sum()

print("Частота значений для 'Payment Mode':")
print(payment_mode_counts)
print("\nСуммирование 'Order Amount' для каждого 'Payment Mode':")
print(payment_mode_sums)
```

```python
#азведочный анализ данных (EDA) (30% от содержания):
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
data['Delivery Time Taken (mins)'].hist(bins=8, edgecolor='black')
plt.title('Распределение времени доставки')
plt.xlabel('Время доставки (минуты)')
plt.ylabel('Частота')
plt.show()


payment_mode_counts = data['Payment Mode'].value_counts()
payment_mode_counts.plot(kind='bar', color='skyblue', edgecolor='black', figsize=(8, 5))
plt.grid(True, linestyle='--', alpha=0.7)
plt.title('Популярность способов оплаты')
plt.xlabel('Способ оплаты')
plt.ylabel('Количество заказов')
plt.show()
```

```python
#Разведочный анализ данных (EDA) (30% от содержания):
import seaborn as sns
import matplotlib.pyplot as plt

correlation_matrix = data[numerical_columns].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", square=True)
plt.title('Тепловая карта корреляции числовых переменных')
plt.show()

```

```python
#Преобразование данных и проектирование характеристик (20% от класса):
def categorize_delivery_time(minutes):
    if minutes <= 20:
        return 'Fast'
    elif minutes <= 40:
        return 'Medium'
    else:
        return 'Slow'

data['Delivery Category'] = data['Delivery Time Taken (mins)'].apply(categorize_delivery_time)
print(data[['Delivery Time Taken (mins)', 'Delivery Category']].head())
```

```python
#Преобразование данных и проектирование характеристик (20% от класса):

data['Avg Item Price'] = data['Order Amount'] / data['Quantity of Items']
print(data[['Avg Item Price']].head())
```

```python
#Преобразование данных и проектирование характеристик (20% от класса):
sorted_data = data.sort_values(by='Order Amount', ascending=False)
print("Топ-5 заказов по сумме:")
print(sorted_data[['Customer Name', 'Order Amount']].head())
```

```python
top_restaurants = data['Restaurant ID'].value_counts().head(10)
print(top_restaurants)
```

```python
delivery_by_restaurant = data.groupby('Restaurant ID')['Delivery Time Taken (mins)'].mean().sort_values()
print(delivery_by_restaurant.head(10))
```

```python
#Преобразование данных и проектирование характеристик (20% от класса):
high_food_rating = data[data['Customer Rating-Food'] >= 4]
print("Заказы с высоким рейтингом еды (>= 4):")
print(high_food_rating)

specific_order = data.loc[data['Order ID'] == 'OD1', ['Order Amount', 'Payment Mode']]
print("Детали заказа 'OD1':")
print(specific_order)
```

```python
#Преобразование данных и проектирование характеристик (20% от класса):
data['Order Hour'] = data['Order Date'].dt.hour

peak_hour_orders = data[(data['Order Hour'] >= 17) & (data['Order Hour'] <= 19)]
print("Заказы во время пиковых часов (17:00-19:00):")
print(peak_hour_orders[['Order ID', 'Order Date', 'Order Hour', 'Order Amount']])

```

```python
pivot_table = data.pivot_table(
    values='Delivery Time Taken (mins)',
    index='Payment Mode',
    aggfunc=['mean', 'count']
)
pivot_table.columns = ['Avg Delivery Time', 'Order Count']
print("Сводная таблица по способам оплаты:")
print(pivot_table)
```

```python
hourly_avg_orders = data.groupby('Order Hour')['Order Amount'].mean()

hourly_avg_orders.plot(kind='bar', figsize=(10, 5), color='skyblue', title='Средняя сумма заказов по часам суток')
plt.xlabel('Час')
plt.ylabel('Средняя сумма заказа')
plt.grid()
plt.show()
```
